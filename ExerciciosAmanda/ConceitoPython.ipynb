### 1- Crie uma classe Aluno que armazene o nome e uma lista de notas. Implemente um  método para calcular a média e outro para verificar se o aluno foi aprovado (média  >= 6). 


```python
class Aluno:
    def __init__(self, nome, notas):
        self.nome = nome
        self.notas = notas  # lista de notas

    def calcular_media(self):
        if len(self.notas) == 0:
            return 0
        return sum(self.notas) / len(self.notas)

    def foi_aprovado(self):
        return self.calcular_media() >= 6


# exemplo de uso:
aluno1 = Aluno("Amanda", [7, 8.5, 8, 9])

print("Aluno:", aluno1.nome)
print("Média:", aluno1.calcular_media())
print("Aprovado?", aluno1.foi_aprovado())
```

### 2- Crie uma classe Conversor que possua um atributo privado taxa_dolar. Implemente  métodos para atualizar a taxa (com validação para não aceitar valores negativos) e  para converter um valor de Real para Dólar. 


```python
class Conversor:
    def __init__(self, taxa_dolar):
        if taxa_dolar < 0:
            raise ValueError("A taxa não pode ser negativa.")
        self.__taxa_dolar = taxa_dolar  # atributo privado

    def atualizar_taxa(self, nova_taxa):
        if nova_taxa < 0:
            raise ValueError("A taxa não pode ser negativa.")
        self.__taxa_dolar = nova_taxa

    def converter_para_dolar(self, valor_em_reais):
        return valor_em_reais / self.__taxa_dolar


# exemplo de uso:
conversor = Conversor(5.20)

print("Convertendo R$ 100 para dólares:", conversor.converter_para_dolar(100))

conversor.atualizar_taxa(5.00)
print("Convertendo R$ 100 com nova taxa:", conversor.converter_para_dolar(100))
```

### 3- Crie as classes Livro e Biblioteca. A classe Biblioteca deve conter uma lista de  objetos do tipo Livro. Implemente métodos para adicionar livros e listar todos os  títulos disponíveis. 


```python
class Livro:
    def __init__(self, titulo):
        self.titulo = titulo


class Biblioteca:
    def __init__(self):
        self.livros = []

    def adicionar_livro(self, livro):
        self.livros.append(livro)

    def listar_livros(self):
        if not self.livros:
            print("Nenhum livro disponível.")
        else:
            print("Livros disponíveis:")
            for livro in self.livros:
                print("-", livro.titulo)


# exemplo de uso:
livro1 = Livro("Harry Potter")
livro2 = Livro("Jogos Vorazes")
livro3 = Livro("Percy Jackson")

biblioteca = Biblioteca()

biblioteca.adicionar_livro(livro1)
biblioteca.adicionar_livro(livro2)
biblioteca.adicionar_livro(livro3)

biblioteca.listar_livros()
```

### 7- Crie uma classe Usuario com login e senha. Implemente um método que valide se  a senha possui pelo menos 8 caracteres e um caractere especial. Se não atender,  o sistema deve solicitar uma nova senha.


```python
# Import necessário
import re

# Definindo a classe Usuario
class Usuario:
    def __init__(self, login, senha):
        self.login = login
        self.senha = senha
    
    # Método para validar senha
    def validar_senha(self):
        """
        Valida se a senha possui pelo menos 8 caracteres
        e pelo menos um caractere especial.
        Retorna True se a senha for válida, False caso contrário.
        """
        # Verifica se a senha tem pelo menos 8 caracteres
        if len(self.senha) < 8:
            return False
        
        # Verifica se a senha contém pelo menos um caractere especial
        # Considerando caracteres especiais como: !@#$%^&*()-_+= etc.
        if not re.search(r"[!@#$%^&*(),.?\":{}|<>]", self.senha):
            return False
        
        return True
    
    # Método para solicitar nova senha se a atual for inválida
    def solicitar_nova_senha(self):
        while not self.validar_senha():
            print("Senha inválida! A senha deve ter pelo menos 8 caracteres e um caractere especial.")
            self.senha = input("Digite uma nova senha: ")
        print("Senha válida!")

# Exemplo de uso
usuario1 = Usuario("joao123", "senha123")  # senha inicial inválida
usuario1.solicitar_nova_senha()
```

### 8- Crie uma classe Camera e uma classe Telefone. Crie uma terceira classe  Smartphone que herde de ambas, demonstrando como o objeto final pode acessar  métodos de as duas classes originais. 


```python
# classe Camera
class Camera:
    def tirar_foto(self):
        print("Foto tirada com sucesso!")

    def gravar_video(self):
        print("Gravando vídeo...")


# classe Telefone
class Telefone:
    def fazer_ligacao(self, numero):
        print(f"Ligando para {numero}...")

    def enviar_mensagem(self, numero, mensagem):
        print(f"Enviando mensagem para {numero}: {mensagem}")


# classe Smartphone herdando de Camera e Telefone
class Smartphone(Camera, Telefone):
    def __init__(self, modelo):
        self.modelo = modelo

    def info(self):
        print(f"Smartphone modelo: {self.modelo}")


# criando objeto Smartphone
meu_celular = Smartphone("XPhone")

# acessando métodos das três classes
meu_celular.info()
meu_celular.tirar_foto()
meu_celular.gravar_video()
meu_celular.fazer_ligacao("1199999-0000")
meu_celular.enviar_mensagem("1199999-0000", "Olá, Amanda!")
```

### 9- Crie uma classe Funcionario que possua um método registrar_entrada(). Utilize o  módulo datetime para capturar o horário exato em que o método foi chamado e  armazene isso em uma lista de logs dentro do objeto. 

```python
from datetime import datetime

class Funcionario:
    def __init__(self, nome):
        self.nome = nome
        self.logs_entrada = []  # lista para armazenar os registros

    def registrar_entrada(self):
        horario_atual = datetime.now()
        self.logs_entrada.append(horario_atual)
        print(f"Entrada registrada para {self.nome} às {horario_atual.strftime('%d/%m/%Y %H:%M:%S')}")

    def mostrar_logs(self):
        print(f"\nRegistros de entrada de {self.nome}:")
        for log in self.logs_entrada:
            print(log.strftime('%d/%m/%Y %H:%M:%S'))


# exemplo de uso
funcionario1 = Funcionario("Amanda")
funcionario1.registrar_entrada()


funcionario1.mostrar_logs()
```

### 10- Crie uma classe Calculadora que contenha apenas métodos estáticos para  operações como soma, subtração e raiz quadrada. O objetivo é usar a classe como  um "namespace" de funções utilitárias sem precisar instanciá-la.

```python
import math

class Calculadora:

    @staticmethod
    def soma(a, b):
        return a + b

    @staticmethod
    def subtracao(a, b):
        return a - b

    @staticmethod
    def raiz_quadrada(numero):
        if numero < 0:
            raise ValueError("Não é possível calcular a raiz quadrada de número negativo.")
        return math.sqrt(numero)


## utilizando a classe sem instanciar
print("Soma:", Calculadora.soma(10, 5))
print("Subtração:", Calculadora.subtracao(10, 5))
print("Raiz Quadrada:", Calculadora.raiz_quadrada(25))
```
